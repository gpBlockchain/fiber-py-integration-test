```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. 获取数据
symbol = "BTC-USD"  # 比特币对美元
data = yf.download(symbol, start="2024-01-01", end="2025-03-13")  # 从2023年到今天
df = data[['Close']].copy()  # 只取收盘价

# 2. 计算短期和长期均线
df['MA5'] = df['Close'].rolling(window=5).mean()  # 5日均线
df['MA20'] = df['Close'].rolling(window=20).mean()  # 20日均线

# 3. 生成交易信号
df['Signal'] = 0  # 默认无信号
df['Signal'][5:] = np.where(df['MA5'][5:] > df['MA20'][5:], 1, 0)  # 金叉=1，死叉=0
df['Position'] = df['Signal'].diff()  # 信号变化：1表示买入，-1表示卖出

# 4. 可视化
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='BTC-USD Close', alpha=0.5)
plt.plot(df['MA5'], label='MA5', alpha=0.8)
plt.plot(df['MA20'], label='MA20', alpha=0.8)

# 标记买入和卖出点
plt.plot(df[df['Position'] == 1].index, df['MA5'][df['Position'] == 1], '^', markersize=10, color='g', label='Buy')
plt.plot(df[df['Position'] == -1].index, df['MA5'][df['Position'] == -1], 'v', markersize=10, color='r', label='Sell')

plt.title('BTC-USD Moving Average Crossover Strategy')
plt.legend()
plt.show()

# 5. 输出交易信号
print("交易信号：")
print(df[['Close', 'MA5', 'MA20', 'Signal', 'Position']].tail(10))
```
